```python
#import dependecies
# Dependencies and Setup
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Reading the CSV source file into pandas via pd.read_csv method
csv_file = "../data/PS4_GamesSales.csv"
PS4_GamesSales= pd.read_csv(csv_file)
PS4_GamesSales
```

# Data Cleaning

```python
# Drop rows with NaNs
PS4_GamesSalesdropna = PS4_GamesSales.dropna(how='any')
PS4_GamesSalesdropna
```

```python
# Drop duplicate rows with the same Game name
PS4_GamesSalesdropdup = PS4_GamesSalesdropna.drop_duplicates("Game", inplace=False)
PS4_GamesSalesdropdup
```

```python
# Drop rows with lots of zeros in the different game sales regions i.e. North America, Europe, Japan, Rest of World
droprowswithlotszeros = PS4_GamesSalesdropdup[(PS4_GamesSalesdropdup['NorthAmerica'] != 0) |\
                                                (PS4_GamesSalesdropdup['Europe'] != 0) &\
                                               (PS4_GamesSalesdropdup['Japan'] != 0) &\
                                                (PS4_GamesSalesdropdup['RestofWorld'] != 0)]
droprowswithlotszeros
```

```python
# Further drop rows with three zeros in the different game sales regions i.e. North America, Europe, Japan, Rest of World
Furtherdrop = droprowswithlotszeros.drop([632, 633, 635, 636, 638, 643, 646, 650, 651, 652, 653, 655, 656, 657, 658,\
                                       663, 664, 669, 671, 674, 676, 678, 685, 694, 697, 699, 704]) 
Furtherdrop
```

```python
# Convert Year column to integer to remove the 0 after the decimal
Cleaned = Furtherdrop.astype({"Year": int})
Cleaned
```

```python
# Export the cleaned dataframe into CSV file
Cleaned.to_csv('./ETLcleaning.csv')
```
